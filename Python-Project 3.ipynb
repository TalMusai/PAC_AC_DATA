## Data Handling

```python
#Checks for missing values (NaN) in each column and returns the total number of missing values per column
happy_df.isnull().sum()
```

```python
#Fills any missing (NaN) values in the DataFrame with the mean of each numeric column
happy_df.fillna(happy_df.mean(numeric_only=True), inplace=True)
```

```python
#Detects outliers in the "Ladder_score" column using the 3 standard deviation rule, and prints them
mean = happy_df["Ladder_score"].mean()
std = happy_df["Ladder_score"].std()
outliers = happy_df[(happy_df["Ladder_score"] > mean + 3*std) | (happy_df["Ladder_score"] < mean - 3*std)]
print(outliers)
```

```python
#Removes the detected outliers from the DataFrame (keeps only values within 3 std from the mean)
happy_df = happy_df[(happy_df["Ladder_score"] <= mean + 3*std) & (happy_df["Ladder_score"] >= mean - 3*std)]
```

## Advanced Analysis

```python
#Groups by region and calculates the mean, count, and median of the happiness score
happy_df.groupby("Regional_indicator")[["Ladder_score"]].agg(["mean", "count", "median"])
```

```python
#Creates a new categorical column based on ranges of the "freedom to make life choices" score
happy_df.loc[:, "Freedom_level"] = pd.cut(
    happy_df["Freedom_to_make_life_choices"],
    bins=[0, 0.3, 0.6, 1],
    labels=["Low", "Medium", "High"]  
)

#Groups by the new freedom level category and shows mean, count, and median of happiness score
happy_df.groupby("Freedom_level", observed=True)[["Ladder_score"]].agg(["mean", "count", "median"])
```

```python
#Creates a new column categorizing countries by happiness levels based on score ranges
happy_df.loc[:, "Happiness_level"] = pd.cut(
    happy_df["Ladder_score"],
    bins=[2.7, 4.5, 5.7, 6.7, 7.8],
    labels=["Low", "Medium", "High", "Very High"]  
)

#Groups by happiness level and shows statistics for GDP per capita
happy_df.groupby("Happiness_level", observed=True)[["Log_GDP_per_capita"]].agg(["mean", "count", "median"])
```

```python
#Categorizes countries into three levels based on healthy life expectancy
happy_df.loc[:, "Life_expectancy_level"] = pd.cut(
    happy_df["Healthy_life_expectancy"],
    bins=3,
    labels=["Low", "Medium", "High"]
)

happy_df.groupby("Life_expectancy_level", observed=True)[["Ladder_score"]].agg(["mean", "count", "median"])
```

```python
# Groups the data by region and freedom level, and calculates the mean, count, and median of happiness scores.
happy_df.groupby(["Regional_indicator", "Freedom_level"], observed=True)[["Ladder_score"]].agg(["mean", "count", "median"])
```

## Visualization

```python
#Pie chart- Distribution of countries by region
plt.figure(figsize=(8, 7))
happy_df["Regional_indicator"].value_counts().plot(
    kind="pie", autopct="%1.1f%%", startangle=90, cmap="Set3", wedgeprops={"edgecolor": "black"}
)
plt.title("Country Distribution by Region", fontsize=14, fontweight="bold")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
#Horizontal bar chart- showing average social support by region
plt.figure(figsize=(10, 6))
happy_df.groupby("Regional_indicator")["Social_support"].mean().sort_values().plot(
    kind="barh", color="skyblue", edgecolor="black"
)
plt.title("Average Social Support by Region", fontsize=14, fontweight="bold")
plt.xlabel("Average Social Support")
plt.ylabel("Region")
plt.tight_layout()
plt.show()
```

```python
#Scatter plot- to visualize the relationship between GDP and life expectancy, grouped by region
plt.figure(figsize=(8, 6))
sb.scatterplot(
    data=happy_df,
    x="Log_GDP_per_capita",
    y="Healthy_life_expectancy",
    hue="Regional_indicator",
    palette="Set2",
    edgecolor="black"
)
plt.title("GDP vs Healthy Life Expectancy by Region", fontsize=14, fontweight="bold")
plt.xlabel("Log GDP per Capita")
plt.ylabel("Healthy Life Expectancy")
plt.legend(title="Region", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#Scatter plot- showing the relationship between happiness and perceived corruption
plt.figure(figsize=(8, 6))
sb.scatterplot(data=happy_df, x="Ladder_score", y="Perceptions_of_corruption", color="steelblue")
plt.title("Ladder Score vs Perceptions of Corruption", fontsize=14, fontweight="bold")
plt.xlabel("Happiness Score (Ladder Score)")
plt.ylabel("Perceptions of Corruption")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
#Regression plot- showing the trend between freedom and generosity, including a best-fit line
plt.figure(figsize=(8,6))
sb.regplot(data=happy_df, x="Freedom_to_make_life_choices", y="Generosity", scatter_kws={"color":"blue"}, line_kws={"color":"red"})
plt.xlabel("Freedom to Make Life Choices")
plt.ylabel("Generosity")
plt.title("Regression Plot: Freedom vs. Generosity", fontsize=14, fontweight='bold')

plt.show()
```

```python
#Heatmap histogram- showing how GDP and life expectancy are distributed together
plt.figure(figsize=(8, 6))
hist = sb.histplot(
    data=happy_df,
    x="Log_GDP_per_capita",
    y="Healthy_life_expectancy",
    cmap="YlGnBu",
    bins=30,
    pthresh=0.05
)
cbar = plt.colorbar(hist.collections[0], ax=hist)
cbar.set_label("Number of Countries", rotation=270, labelpad=15)

plt.title("Density Plot: GDP vs Healthy Life Expectancy", fontsize=14, fontweight="bold")
plt.xlabel("Log GDP per Capita")
plt.ylabel("Healthy Life Expectancy")
plt.tight_layout()
plt.show()
```

```python
#Histogram with KDE- showing the distribution of happiness scores
plt.figure(figsize=(8, 6))
sb.histplot(happy_df["Ladder_score"], bins=20, kde=True, color="orange")
plt.title("Distribution_of_Happiness_Scores", fontsize=14, fontweight='bold')
plt.xlabel("Ladder_Score")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# Line plot- Comparison of Ladder Score among the top 10 happiest countries in 2024
top_countries = happy_df.head(10)
plt.figure(figsize=(12, 6))
plt.plot(top_countries["Country_name"], top_countries["Ladder_score"],
         marker='o', linestyle='-', linewidth=2)

plt.title("Ladder Score of the Top 10 Happiest Countries – 2024", fontsize=16, fontweight='bold')
plt.xlabel("Country", fontsize=12)
plt.ylabel("Ladder Score", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(["Ladder Score"], loc='upper right')

plt.tight_layout()
plt.show()
```

```python
#Box plot- showing how happiness varies across different freedom levels
plt.figure(figsize=(10, 6))
sb.boxplot(
    data=happy_df,
    x="Freedom_level",
    y="Ladder_score",
    hue="Freedom_level",  # הצבע לפי אותו ערך כמו ב-x
    palette={"Low": "#FF9999", "Medium": "#FFCC99", "High": "#99CC99"},
    legend=False  # נבטל את המקרא כי הוא כפול
)

plt.title("Happiness Score Distribution by Freedom Level", fontsize=14, fontweight='bold')
plt.xlabel("Freedom to Make Life Choices (Binned)", fontsize=12)
plt.ylabel("Happiness Score (Ladder Score)", fontsize=12)
plt.tight_layout()
plt.show()
```
